```python
image = Image.open("raccoon_dataset/images/raccoon-10.jpg") # A random image

image
```

```python
# I create transformers with different characteristics, you can change as you want

image_transforms = {

    't1': transforms.Compose([
        transforms.ToTensor(),
    ]),
    't2': transforms.Compose([
        transforms.RandomHorizontalFlip(),
        transforms.GaussianBlur(kernel_size=(11, 11)),
        transforms.ToTensor(),
    ]),
    't3': transforms.Compose([
        transforms.RandomResizedCrop(size=(250, 250)),
        transforms.ToTensor(),
    ]),
    't4': transforms.Compose([
        transforms.RandomHorizontalFlip(),
        transforms.RandomCrop((250, 250)),
        transforms.ToTensor(),
    ]),
    't5': transforms.Compose([
        transforms.RandomRotation(degrees=30),
        transforms.ToTensor(),
    ]),
}
```

```python
# display images one by one with their title
# their titles indicate which transformer they are manufactured by

import numpy
from PIL import Image
from matplotlib import pyplot as plt

for transform in ['t1', 't2', 't3', 't4', 't5']:
    new_img =  image_transforms[transform](image)

    plt.title(label=transform)
    plt.imshow(new_img.permute(1,2,0), cmap='gray')
    plt.show()

    print("\n")
```
